```python
def OLS(i):
    mapo=pd.read_excel("mapo_20"+str(i)+".xlsx")
    t=pd.merge(mapo,x_values,on="단지명",how='left')
    t["층"]=t["층"].apply(lambda x: 0 if x<=0 else x)
    t["상대층수"]=t["층"]/t["전체층수"]
    x=t[["전용면적(㎡)","건축년도","거리_미터","브랜드","상대층수"]]
    #x=t[["전용면적(㎡)","건축년도","거리_미터","브랜드","층"]]
    x=sm.add_constant(x)
    y=t[["거래금액(만원)"]]
    model=sm.OLS(y,x)
    results=model.fit()
    #results=model.fit(cov_type="HC1")
    return results.summary()
```

## 잡

```python
mapo=pd.read_excel("mapo_2020.xlsx")
t=pd.merge(mapo,x_values,on="단지명",how='left')
t["층"]=t["층"].apply(lambda x: 0 if x<=0 else x)
t["상대층수"]=t["층"]/t["전체층수"]
```

```python
from sklearn.utils import shuffle
t=shuffle(t)
t.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split

x=t[["전용면적(㎡)","건축년도","거리_미터","브랜드","상대층수"]]
y=t[["거래금액(만원)"]]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)

mlr = LinearRegression()
mlr.fit(x_train, y_train) 

y_predict=mlr.predict(x_test)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_predict, alpha=0.4)
plt.xlabel('Actual')
plt.ylabel("Predict")
plt.title("y_predict")
plt.show()
```

```python
print(mlr.score(x_train,y_train))
```

```python
from sklearn.model_selection import train_test_split

x=t[["전용면적(㎡)","건축년도","거리_미터","브랜드","상대층수","시군구"]]
x=pd.get_dummies(x)
y=t[["거래금액(만원)"]]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)

mlr = LinearRegression()
mlr.fit(x_train, y_train) 

y_predict=mlr.predict(x_test)

plt.scatter(y_test, y_predict, alpha=0.4)
plt.xlabel('Actual')
plt.ylabel("Predict")
plt.title("y_predict")
plt.show()
```

```python
print(mlr.score(x_train,y_train))
```

## 지역변수

```python
with_locals=t[["전용면적(㎡)","건축년도","거리_미터","브랜드","상대층수","시군구"]]
with_locals=pd.get_dummies(with_locals)
with_locals
```

```python
x=with_locals
y=t[["거래금액(만원)"]]
x=sm.add_constant(x)
model=sm.OLS(y,x)
results=model.fit()
results.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

x_set=t[["전용면적(㎡)","건축년도","거리_미터","브랜드","층"]]

vif=pd.DataFrame()
vif["VIF Factor"]=[variance_inflation_factor(x_set.values, i) for i in range(x_set.shape[1])]
vif["features"]=x_set.columns
vif
```
